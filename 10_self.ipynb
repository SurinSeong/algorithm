## 종료조건이 있는 재귀함수

### 함수

```python
def openBox():
    global count
    print("Open the box.")
    count -= 1
    if count == 0:
        print("Gift!")
        return
    openBox()
    print("Close the box.")
```

### 메인

```python
count = 10
openBox()
```

## 숫자 합계 내기

### 함수

```python
def addNumber(num):
    if num == 1:
        return 1
    return num + addNumber(num-1)
```

### 메인

```python
print(addNumber(100))
```

## 구구단 출력하기

### 함수

```python
# 하나의 단 구하기
def gugu(d, n):
    print(f"{d} x {n} = {d*n}")
    if n < 9:
        gugu(d, n+1)
```

### 메인

```python
for dan in range(2, 10):
    print(f"--- {dan}단 ---")
    gugu(dan, 1)
```

## 별 모양 출력

### 함수

```python
def printStar(num):
    if num > 0:
        printStar(num-1)
        print("★" * num)
```

### 메인

```python
printStar(5)
```
