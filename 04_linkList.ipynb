```python
## 함수
# 노드 자료형 만들기
class Node():
    def __init__(self):
        self.data = None
        self.link = None

# 노드 순서대로 출력 함수
def printNodes(start):
    current = start
    print(current.data, end=" ")
    while (current.link != None):
        current = current.link
        print(current.data, end=" ")
    print()
    
# 노드 삽입 함수 : 찾는 노드 앞에 새로운 노드 추가
def insert_node(find_data, insert_data):
    global memory, head, pre, current
    # case 1 : 하필 첫 번째 앞에 삽입 (찾는 노드가 첫 번째 노드)
    if (find_data == head.data):
        node = Node()
        node.data = insert_data
        # 새로운 노드의 링크는 head가 가리키는 노드
        node.link = head
        # head 노드를 새로운 노드로 지정
        head = node
        memory.append(node) # 안 중요
        return
    # case 2 : 찾는 노드가 첫 번째 노드가 아니다.
    # 첫 번째부터 찾기 시작
    current = head
    while (current.link != None):
        pre = current
        current = current.link
        # 현재의 노드가 찾는 노드이면
        if (current.data == find_data):
            # 새로운 노드 생성
            node = Node()
            node.data = insert_data
            # 새로운 노드의 링크는 current가 가리키는 노드
            node.link = current
            # pre의 링크는 새로운 노드를 가리킴
            pre.link = node
            memory.append(node)
            return
    # case 3 : 찾는 노드가 없어 맨 뒤에 추가
    # 빈 노드 만들기
    node = Node()
    node.data = insert_data
    current.link = node
    memory.append(node)
    return

# 노드 삭제 함수
def delete_node(delete_data):
    global memory, head, pre, current
    # case 1 : 하필 머리 삭제 (다현)
    if (delete_data == head.data):
        current = head
        head = head.link
        del(current)
        return
    # case 2 : 중간 노드 삭제
    current = head
    while (current.link != None):
        pre = current
        current = current.link
        if (current.data == delete_data):
            pre.link = current.link
            del(current)
            return
    # case 3 : 삭제할 노드 찾지 못했을 때
    return

# 노드 검색
def find_node(find_data):
    global memory, head, pre, current
    current = head
    if (find_data == current.data):
        return current
    while (current.link != None):
        current = current.link
        if (find_data == current.data):
            return current
    return Node()
```

```python
## 변수
# 앞으로 만들 노드 넣어두는 곳
memory = []
# 변수 준비
head, pre, current = None, None, None

# 내가 사용할 데이터
data_array = ["다현", "정연", "쯔위", "사나", "지효"]
```

```python
## 메인
# 첫 번째 노드 지정
node = Node()
node.data = data_array[0]
head = node
memory.append(node) # 생략가능

# 첫 번째를 제외한 나머지 노드는 반복문으로 처리
for data in data_array[1:]:
    pre = node
    # 빈 노드 생성
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

# 출력
# current = head
# print(current.data, end=" ")
# while (current.link != None):
#     current = current.link
#     print(current.data, end=" ")
# print()

printNodes(head)

# insert_node("다현", "화사")
# insert_node("사나", "솔라")
# insert_node("재남", "문별")
# printNodes(head)

# 노드 삭제
# delete_node("다현")
# delete_node("쯔위")
# printNodes(head)

retNode = find_node("사나")
print(retNode.data, "의 뮤비 시작 !")
```
